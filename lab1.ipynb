## Задание

Необходимо вычислить значение выражения:
$$Z=\frac{x^y-\sqrt[3]{x}}{a\cdot({k\_to + 120\cdot|x-y|})}$$
Для каждой переменной необходимо вывести ее значение и тип.

## Анализ выражения и определение переменных

Проанализируем выражение и определим переменные, которые потребуются для выполнения задания. Независимые переменные:  x,y,a,k_to . Переменная, которая должна быть вычислена на основе независимых:  Z .

````python
x=10
y=30.67
a=5
k_to=5
print("x = ", x, ", type(x) = ", type(x))
print("y = ", y, ", type(y) = ", type(y))
print("k_to = ", k_to, ", type(k_to) = ", type(k_to))
print("a = ", a, ", type(a) = ", type(a))
````

## Вычисление выражение, печать результатов

Код для вычисления значения выражения $Z$:

````python
Z=(x**y-x**(1/3))/(a*(k_to+120*abs(x-y)))
print('Z=',Z)
print('Type of Z: ', type(Z))
````

Основные функции, используемые в данной практической работе: print(), type(), abs()

результат вычисления выражения в соответствие с вариантом:

ВАРИАНТ 4 $$c = gh + b \cdot q3 - x + \frac{y}{w}$$

````python
# объявление переменных с начальными значениями разных типов
g = 2.5      # float
h = 4        # int
b = 10       # int
q = 3        # int
x = 15.7     # float
y = 100      # int
w = 20.0     # float
#вывод значений переменных с использованием форматной строки
print(f"g = {g}, type: {type(g)}")
print(f"h = {h}, type: {type(h)}")
print(f"b = {b}, type: {type(b)}")
print(f"q = {q}, type: {type(q)}")
print(f"x = {x}, type: {type(x)}")
print(f"y = {y}, type: {type(y)}")
print(f"w = {w}, type: {type(w)}")
# вычисление выражения из вариант 4
c = g * h + b * (q**3) - x + y / w
# вывод результата
print(f"\nРезультат вычисления:")
print(f"c = g*h + b*q³ - x + y/w")
print(f"c = {g}*{h} + {b}*{q}³ - {x} + {y}/{w}")
print(f"c = {g*h} + {b*(q**3)} - {x} + {y/w}")
print(f"c = {c}")
print(f"Тип результата: {type(c)}")
````
