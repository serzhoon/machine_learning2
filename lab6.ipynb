Реализуем фильтр для выделения горизонтальных краев на изображении. Этот фильтр будет реагировать на горизонтальные переходы яркости, что полезно для анализа горизонтальных структур.

```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np
img = Image.open('cat.jpeg').convert('L') # загрузка изображения
sobel_horizontal_kernel = ImageFilter.Kernel(
    size=(3, 3),
    kernel=[
        -1, 0, 1,
        -2, 0, 2,
        -1, 0, 1
    ],
    scale=1,
    offset=0
)
edges_img = img.filter(sobel_horizontal_kernel)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Оригинал (серый)')
plt.imshow(img, cmap='gray')
plt.subplot(1, 2, 2)
plt.title('Горизонтальные края (Sobel)')
plt.imshow(edges_img, cmap='gray')
plt.show()
```

2. **Поясните назначение метода convert(). Поясните возможные значения
параметра данного метода.**<BR>

Метод convert() используется для изменения цветового режима изображения. Он позволяет переводить изображение из одного формата в другой, например, в оттенки серого, в черно-белое или в формат CMYK для печати.

Возможные значения параметра mode:

'1' — бинарное черно-белое

'L' — оттенки серого (8 бит)

'P' — индексированное изображение с палитрой

'RGB' — полноцветное изображение без прозрачности

'RGBA' — RGB + альфа-канал (прозрачность)

'CMYK' — цвета для полиграфии

'HSV', 'HSL' — цветовые модели тон-насыщенность-яркость

'YCbCr' — используется в JPEG

'I', 'F' — целые и вещественные числа для специальных задач

3. **Что такое фильтрация изображений? Приведите примеры практического

Фильтрация изображений — это преобразование изображения для улучшения качества, выделения деталей или подготовки к анализу. Выполняется с помощью математических операций, например, свёртки.Применяются в фотографии, медицине, компьютерном зрении, дизайне и промышленности.

Примеры фильтров и их применения:

BLUR — размытие (для анонимизации).

SHARPEN — повышение резкости (улучшение фото).

FIND_EDGES — выделение границ (в машинном зрении).

EMBOSS — эффект рельефа (дизайн).

CONTOUR — контурное изображение (графика).

**4. Поясните формат изображения RGB**

Формат изображения RGB (Red, Green, Blue) представляет цвета как комбинацию красного, зеленого и синего каналов, каждый из которых может принимать значения от 0 до 255. Это позволяет создавать широкий спектр цветов.

5. **Поясните значение терминов: растр, формат изображения, палитра,
глубина, разрешение изображения.**

Растр -  Способ представления изображения в виде сетки пикселей.

Формат изображения - Способ хранения и кодирования данных изображения (например, JPEG, PNG).

Палитра - Набор цветов, используемых в изображении (обычно в индексированных изображениях).

Глубина - Количество битов, используемых для представления каждого пикселя (например, 8-битная глубина).

Разрешение изображения - Количество пикселей в изображении (ширина x высота).
