Разработать класс для представления объекта «Прямоугольный параллелепипед». Реализуйте все необходимые поля данных (закрытые) и методы позволяющие:
* создавать объекты-параллелепипеды;

## Решение

Решение данной задачи состоит из следующих этапов:
проектирование класса; 2) объявление класса (кодирование); 3)

1.   проектирование класса
2.   объявление класса (кодирование)
3.   демонстрация использования класса (создание объекта и вызов методов)

Проектирование класса можно производить с использованием любого средства проектирования (UML-редактор, онлайн средства построения диаграмм ([https://www.draw.io](https://www.draw.io)), обычный графический редактор). На рисунке показана диаграмма классов для спроектированного класса Par.


На данном этапе проектирование завершается. Переходим к объявлению класса `Par`.

````python
class Par:
  # Объявляем конструктор класса
  def __init__(self, pa, pb, pc):
    self.a=pa
    self.b=pb
    self.c=pc

  # Объявляем метод для вычисления площади поверхности фигуры
  def GetS(self):
    return 2*(self.a*self.b + self.b*self.c + self.a*self.c)

  # Объявляем метод для вычисления объема
  def GetV(self):
    return self.a*self.b*self.c

  # Объявление метода для вывода информации о фигуре
  def Info(self):
    print('*'*20)
    print('Прямоугольный параллелепипед: \n')
    print('Размеры: %.2f x %.2f x %.2f' % (self.a, self.b, self.c))
    print('Площадь: %.3f' % self.GetS())
    print('Объем: %.3f' % self.GetV())
    print('*'*20)

  def Load(path_to_file):
    # Предполагаем, что в файле записаны 3 числа в строку через пробел
    d = open(path_to_file).read().strip().split()
    tmp = Par(float(d[0]), float(d[0]), float(d[0]))
    return tmp
````

В данном листинге производится объявление типа данных `Par` с использованием зарезервированного слова `class`. Внутри класса объявляются функции (методы класса) и переменные (поля данных класса). Следует обратить внимание на использование зарезервированного слова `self`.
Также внутри класса объявлен метод `__init__()`. Данный метод имеет специальное назначение - он вызывается в момент создания переменной данного класса автоматически. Данный метод называется конструктор класса.

Использование разработанного типа данных `Par`

````python
# Создание объекта p1 типа Par (используется конструктор)
p1 = Par(10, 20, 30)
# Вызов метода класса
p1.Info()
````

Это значит, что метод не принадлежит конкретному экземпляру (объекту), он принадлежит самому классу. Такие методы называют статическими. Для вызова статического метода нет необходимости явно создавать объект класса, используя конструктор. Статический метод вызывается из класса.

## Вариант 4
Класс «Точка в пространстве». Реализовать ввод и вывод полей
данных, вычисление расстояния до введенной пользователем точки,
расстояния от начала координат, а также вывод информации об
объекте.  

````python
class Point3D:
  def __init__(self, x, y, z):
      self.__x = x
      self.__y = y
      self.__z = z

  # статический метод для ввода точки с клавиатуры
  @staticmethod
  def input_point():
      x = float(input("Введите координату x: "))  # ввод X
      y = float(input("Введите координату y: "))  # ввод Y
      z = float(input("Введите координату z: "))  # ввод Z
      return Point3D(x, y, z)
  # метод для вывода информации о точке
  def print_info(self):
      print(f"Точка в пространстве: ({self.__x:.2f}, {self.__y:.2f}, {self.__z:.2f})")
  # метод для вычисления расстояния до другой точки
  def distance_to(self, other):
      dx = self.__x - other.__x
      dy = self.__y - other.__y
      dz = self.__z - other.__z
 # формула расстояния в 3D:корень из суммы квадратов разностей
      return (dx**2 + dy**2 + dz**2) ** 0.5

  # метод для вычисления расстояния от начала координат
  def distance_from_origin(self):
      return (self.__x**2 + self.__y**2 + self.__z**2) ** 0.5
  # метод для расчета расстояния до точки, введенной пользователем
  def distance_to_user_input(self):
      print("Введите координаты точки для расчета расстояния:")
      other = Point3D.input_point()
      return self.distance_to(other)

if __name__ == "__main__":

  point1 = Point3D(2, 3, 4)
  point1.print_info()

  print("Расстояние от начала координат:", point1.distance_from_origin())

  point2 = Point3D.input_point()
  point2.print_info()

  print("Расстояние между точками:", point1.distance_to(point2))

  distance = point1.distance_to_user_input()
  print("Расстояние до введенной точки:", distance)
````

1. **Дайте определение понятию класс. <br>**
Класс — это шаблон или чертеж, на основе которого создаются объекты (экземпляры). Он определяет набор данных (полей/атрибутов) и поведения (методов), которые будут характерны для всех объектов этого типа. Класс инкапсулирует данные и функциональность, связанную с этими данными, что является основой объектно-ориентированного программирования.

2. Продемонстрируйте использование метода __init__. Поясните назначение
данного метода. <br>
Метод __init__ — это конструктор класса. Он автоматически вызывается при создании нового объекта и используется для инициализации его атрибутов.


```
 class Example:
    def __init__(self, value):
        self.value = value  # Инициализация атрибута

obj = Example(10)  # Создание объекта, вызов __init__
print(obj.value)   # Вывод: 10
```




3.  Поясните назначение статических методов и их отличия от методов класса. <br>
Статические методы принадлежат классу, а не его экземплярам. Они не имеют доступа к self и не могут изменять состояние объекта. Объявляются с помощью декоратора @staticmethod.<Br>
Отличия от методов класса:

Методы класса (без @staticmethod) имеют доступ к self (экземпляру) и могут изменять его состояние.

Статические методы используются для операций, не зависящих от состояния объекта.

4. Для чего используется параметр self в каждом методе класса?<br>
self — это ссылка на текущий экземпляр класса. Он автоматически передается в методы класса и позволяет обращаться к атрибутам и методам объекта.

5. Сколько методов __init__ может быть объявлено внутри класса?<br>
В классе может быть объявлен только один метод __init__. Это специальный метод, который Python интерпретирует как конструктор. Попытка объявить второй __init__ приведет к переопределению первого.
